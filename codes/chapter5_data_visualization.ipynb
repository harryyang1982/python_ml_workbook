## 1. Matplotlib

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots() # (1) figure axes  

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

ax.plot(X_1, Y_1) # (2) plot    
ax.set(title='cos graph', # (3)  , X , Y  
       xlabel='X',
       ylabel='Y');

plt.show() # (4)  
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1) #  1,  2   
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
ax1 = plt.subplot(321)
plt.plot(x, y_1)
ax2 = plt.subplot(322)
plt.plot(x, y_1)
ax3 = plt.subplot(312)
plt.plot(x, y_1)
ax4 = plt.subplot(325)
plt.plot(x, y_1)
ax5 = plt.subplot(326)
plt.plot(x, y_1)

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")

plt.show()
```

```python
plt.plot(X_1, Y_1, c="b",
         linestyle='dashed')
plt.plot(X_2, Y_2, c="r",
         ls="dotted")

plt.show()
```

### 

```python
plt.plot(X_1, Y_1, color='b',
         ls='dashed')
plt.plot(X_2, Y_2, color='r',
         ls='dotted')

plt.title('Two Lines')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c='b')
ax_1.set_title('Figure 1')
ax_2.plot(X_2, Y_2, c='g')
ax_2.set_title('Figure 2')

plt.show()
```

```python
plt.plot(X_1, Y_1,
         color='b',
         linestyle='dashed',
         label='line_1')
plt.plot(X_2, Y_2,
         color='r',
         linestyle='dotted',
         label='line_2')
plt.legend(
    shadow=True,
    fancybox=False,
    loc="upper right"
)

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N) ** 2)
plt.scatter(x, y,
            s=area, c=colors,
            alpha=0.5)
plt.show()
```

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0, 8, 2)

plt.bar(X + 0.00, data[0], color = 'b', width= 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width= 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)

plt.xticks(X+0.50, ("A", "B", "C", "D"))

plt.show()
```

###   

```python
data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ['A', 'B', 'C']
X = np.arange(data.shape[1])

for i in range(3):
    plt.bar(X, data[i],
            bottom = np.sum(
                data[:i], axis=0),
            color = color_list[i],
            label=data_label[i])
    
plt.legend()
plt.show()
```

## 2. Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

fmri = sns.load_dataset('fmri')
sns.set_style('whitegrid')
sns.lineplot(x='timepoint', y='signal', data=fmri)
```

```python
fmri.sample(n=10, random_state=1)
```

```python
sns.lineplot(x='timepoint', y='signal', hue='event', data=fmri)
```

### Regression graph

```python
tips = sns.load_dataset('tips')
sns.regplot(x='total_bill', y='tip', data=tips, x_ci=95)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

```python
tips
```

### counter plot

```python
sns.countplot(x='smoker', hue='time', data=tips)
```

###  

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
tips
```

### violin plot and swarm plot

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
```

```python
sns.swarmplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
```

### facet_grid

```python
g = sns.FacetGrid(tips, col='time', row='sex')
```

```python
g = sns.FacetGrid(tips, col='time', row='sex')

g.map(sns.scatterplot,
      'total_bill', 'tip')
```

```python
g = sns.FacetGrid(tips, col='time', row='sex')

g.map_dataframe(sns.histplot,
                x='total_bill')
```

## 2. plotly

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length')
fig.show()
```

```python
fig=px.parallel_coordinates(df, color='species_id',
                            labels={'species_id': 'Species',
                                    'sepal_width':'Sepal Width', 'sepal_length': 'Sepal Length',
                                    'petal_width': 'Petal Width',
                                    'petal_length': 'Petal Length', },
                            color_continuous_scale = px.colors.diverging.Tealrose,
                            color_continuous_midpoint=2)
fig.show()
```
